# Exploratory Data Analysis (EDA) on Electric Vehicle Analytics Dataset  

Exploratory Data Analysis (EDA) is the process of examining and visualizing a dataset  
to understand its structure, patterns, and relationships before applying modeling or  
statistical techniques.  

The dataset `electric_vehicle_analytics.csv` contains detailed records related to **electric vehicles (EVs)**.  
It includes various attributes such as vehicle specifications, usage metrics, and performance factors.  

The objective of this EDA is to:  
- Review and clean the dataset  
- Summarize and visualize numerical and categorical features  
- Identify key patterns and correlations in EV attributes  
- Gain insights that can support decision-making in the electric vehicle ecosystem

Outlier Detection

```python
#1. Detect outliers in Range_km using IQR method
Q1 = df["Range_km"].quantile(0.25)
Q3 = df["Range_km"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_range = df[(df["Range_km"] < lower_bound) | (df["Range_km"] > upper_bound)]
print("Number of outliers in Range:", outliers_range.shape[0])
outliers_range[["Make","Model","Range_km"]].head()
```

```python
# 2. Boxplot to visualize outliers in Monthly_charging_Cost_USD
plt.figure(figsize=(6,5))
sns.boxplot(y=df["Monthly_Charging_Cost_USD"], color="orange")
plt.title("Outlier Detection in Monthly Charging Cost (USD)")
plt.ylabel("Monthly Charging Cost (USD)")
plt.show()
```

```python
#3. Outlier detection across multiple features (visual
plt.figure(figsize=(12,6))
sns.boxplot(data=df[["Battery_Capacity_kWh",
                     "Battery_Health_%",
                     "Range_km",
                     "Monthly_Charging_Cost_USD",
                     "Resale_Value_USD"]],
            palette="Set2")

plt.title("Outlier Detection across Key Features")
plt.xlabel("Features")
plt.ylabel("Values")
plt.xticks(rotation=30)   # âœ… rotate feature names
plt.show()
```
